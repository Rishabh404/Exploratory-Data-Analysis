```python
# Import the `libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
# Load in the data with `read_csv()`
digits = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/optdigits/optdigits.tra",
                     header=None)

iris = pd.read_csv('Iris.csv')
```

```python
# Print out `digits`
print(digits)
```

```python
# This function gives information about mean, median, minimum and maximum values of quantity of data excluding NaN values

digits.describe()
```

```python
#head() - first 5 rows of dataframe, tail() - last 5 rows of dataframe

# Inspect the first 5 rows of `digits`
first = digits.head()

# Inspect the last 5 rows
last = digits.tail()
```

```python
"""
Sampling the dataset

In case of a large dataset, it is recommended to take sample of it to get the feel of the data quickly. There are
two ways to do it. One involves making use of sample() function by inputing in the number of rows for which we need
the sample as parameter. Other method makes use of random package which has the modules sample(), range() and len()
and combination of these might be used to get the requisite sample of the dataset.
                
"""
```

```python
# Using sample() function

x = digits.sample(5)
print(x)
```

```python
# Using sample(), range() and len() functions included in random package

# import sample from random
from random import sample

# Create a random index
randomIndex = np.array(sample(range(len(digits)), 5))

# Get 5 random rows
digitsSample = digits.loc[randomIndex]

# Print the sample
print(digitsSample)
```

```python
# Further inspection of data with query() function

# Petal length greater than sepal length?
iris.query('PetalLengthCm > SepalLengthCm')

```

```python
# Petal length equals sepal length?
iris.query('PetalLengthCm == SepalLengthCm')
```

```python
# Petal length less than sepal length?
iris.query('PetalLengthCm < SepalLengthCm')
```
